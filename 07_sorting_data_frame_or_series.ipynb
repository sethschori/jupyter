```python
import pandas as pd
```

```python
data_file = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
```

```python
movies = pd.read_csv(data_file)
```

```python
movies.head()
```

## The new way of sorting (pandas >= 0.17)

### Sort a Series

```python
movies.title.sort_values()
```

```python
type(movies.title.sort_values())  # returns a Series, which tells us that .sort_values() is a Series method
```

```python
# sort in descending order
movies.title.sort_values(ascending=False)
```

```python
# .sort_values() doesn't change the underlying order.
movies.title
```

### Sort a DataFrame by a Series
This is a DataFrame method that returns a DataFrame. It does not change the underlying sort order.

```python
movies.sort_values('title')
```

### Sort a DataFrame by multiple columns

```python
movies.sort_values(['content_rating', 'duration'])
```
